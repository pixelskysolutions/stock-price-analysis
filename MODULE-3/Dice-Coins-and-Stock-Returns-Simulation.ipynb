```python
# Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# ðŸŽ² Module 3 - Project 1: Dice, Coins, and Stock Returns Simulation

# Probability is the foundation of finance.  
# In this project, weâ€™ll:
# 1. Simulate **coin tosses** and estimate probabilities.
# 2. Simulate **dice rolls** and explore probability distributions.
# 3. Apply probability to **financial returns** with stock simulations.

# This builds intuition for randomness and uncertainty in finance.
```

```python
# ðŸª™ Step 1: Coin Toss Simulation
```

```python
## ðŸ”Ž Explanation
# - A fair coin has 50% chance for Heads, 50% for Tails.
# - With enough simulations (10,000), the frequencies converge close to 0.5 each.
# - This is the **Law of Large Numbers** in action.
```

```python
# Simulate 10,000 coin flips
np.random.seed(42)
coin_flips = np.random.choice(["Heads", "Tails"], size=10_000)

# Count frequencies
coin_counts = pd.Series(coin_flips).value_counts(normalize=True)

print("Coin Toss Probabilities:")
print(coin_counts)

# Plot
coin_counts.plot(kind="bar", title="Coin Toss Simulation (10,000 flips)")
plt.ylabel("Probability")
plt.show()
```

```python
# ðŸŽ² Step 2: Dice Roll Simulation
```

```python
### ðŸ”Ž Explanation
# - A fair die has 6 outcomes (1â€“6), each with 1/6 â‰ˆ 16.7% probability.
# - As with coin flips, with many rolls the empirical distribution converges to theory.
```

```python
# Simulate 10,000 dice rolls
dice_rolls = np.random.randint(1, 7, size=10_000)

# Count probabilities
dice_counts = pd.Series(dice_rolls).value_counts(normalize=True).sort_index()

print("Dice Roll Probabilities:")
print(dice_counts)

# Plot
dice_counts.plot(kind="bar", title="Dice Roll Simulation (10,000 rolls)")
plt.ylabel("Probability")
plt.show()
```

```python
# Step 3: Simulating Stock Returns (1 year)
```

```python
# Parameters
np.random.seed(42)
n_days = 252  # ~1 year of trading days
mu = 0.0005   # expected daily return (~12% annually)
sigma = 0.02  # daily volatility (~32% annually)

# Simulate daily returns (Normal distribution)
daily_returns = np.random.normal(mu, sigma, n_days)

# Convert to price series, starting at 100
price = 100 * (1 + daily_returns).cumprod()

# Put into DataFrame
stock_df = pd.DataFrame({
    "Daily Return": daily_returns,
    "Stock Price": price
})

stock_df.head()
```

```python
# ðŸ“Š Step 4: Visualizing Stock Returns
```

```python
### ðŸ”Ž Explanation
# - We assume returns follow a **normal distribution** with mean (Î¼) and standard deviation (Ïƒ).
# - Over time, randomness accumulates, and the price path evolves unpredictably.
# - This randomness is what makes finance a game of probabilities â€” not certainties.
```

```python
# Plot stock price simulation
plt.figure(figsize=(10,5))
plt.plot(stock_df["Stock Price"], label="Simulated Stock Price")
plt.title("ðŸ“ˆ Simulated Stock Price (1 Year)")
plt.xlabel("Day")
plt.ylabel("Price")
plt.legend()
plt.show()

# Plot distribution of returns
sns.histplot(stock_df["Daily Return"], bins=50, kde=True)
plt.title("Distribution of Daily Returns")
plt.xlabel("Daily Return")
plt.show()
```
